```python
from datetime import datetime
import re
import pandas as pd
```

```python
fidPath = "LINE_record.txt"
# Creat a empty List
text    = []

# Open the txtfile  
with open(fidPath, "r", encoding = "utf-8") as file:
    while True:
        lines = file.readline() # Read the line from the txt file
        if not lines:
            break               # Read the line until the end of the txt file        
        lines = lines.split()
        if lines != []:
            text.append(lines)  # Save contents as a list         
```

```python
time_memembers = []
members = []
membersName = []

for ii in range(0, len(text)):
    boolValue = "ä¸Šåˆ" in text[ii][0] or "ä¸‹åˆ" in text[ii][0]
    if len(text[ii]) >= 3 and boolValue:
        time_memembers.append([text[ii][0], text[ii][1], ii])        
        members.append(text[ii][1])

# Remove noises in lists
members = [x for x in members if not ("é‚€è«‹" in x)]
members = [x for x in members if not ("å·²å°‡" in x)] 

# Remove duplicates in the "members" list
membersName = list(set(members))

for ii in range(0,len(time_memembers)):
    if "ä¸‹åˆ" in time_memembers[ii][0]:
        timeTemp = datetime.strptime(time_memembers[ii][0].replace("ä¸‹åˆ","PM "), "%p %I:%M")
        time_memembers[ii][0] = timeTemp.strftime("%H:%M")
    elif "ä¸Šåˆ" in time_memembers[ii][0]:
        timeTemp = datetime.strptime(time_memembers[ii][0].replace("ä¸Šåˆ","AM "), "%p %I:%M")
        time_memembers[ii][0] = timeTemp.strftime("%H:%M")
        
# Find the date of the conversation record 
dateTime = []
for ii in range(0, len(text)):
    match = re.search(r'\d{4}/\d{2}/\d{2}', text[ii][0])
    # \d  match any number
    # {4} repeated exactly 4 times
    
    if match is not None:
        dateTime.append([match.group(0),ii])

# Add a ghost date
dateTime.append(['9999/99/99', len(time_memembers) * 10]) 

dateIndex = 1 #ã€€Start from the index 1
for ii in range(0, len(time_memembers)):
    if time_memembers[ii][2] < dateTime[dateIndex][1]:
        time_memembers[ii][0] = dateTime[dateIndex - 1][0] + ' ' + time_memembers[ii][0]
    else:
        time_memembers[ii][0] = dateTime[dateIndex][0] + ' ' + time_memembers[ii][0]
        dateIndex += 1  
```

```python
commentCountList = []
for ii in range(0, len(membersName)):
    countNum = members.count(membersName[ii])
            
    for sublist in time_memembers:
        if membersName[ii] in sublist:
            updateTime = sublist[0]
        
    dictnary = {"name" : membersName[ii], 
                "number of posting comments" : countNum,
                "last update" : updateTime}
    commentCountList.append(dictnary)
```

```python
# Create the dataframe
dF = pd.DataFrame(commentCountList)
dF = dF[["name", "number of posting comments", "last update"]]
dF.index.name = "No."
dF.to_excel("chatroomAna.xlsx", index = False)
```
